```python
import torch
import torch.nn as nn
import torchvision.models as models

# Define the model architecture
model = models.efficientnet_b0(pretrained=False)  # Load model without pretrained weights
model.classifier[1] = nn.Linear(model.classifier[1].in_features, 2)  # Match the output classes (normal, porn)

# Load the trained model weights
model.load_state_dict(torch.load("efficientnet_b0_v2.pth", map_location=torch.device("cpu")))  # Use "cuda" if needed
model.eval()  # Set to evaluation mode
```

```python
from PIL import Image
import torchvision.transforms as transforms

# Define transformation
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
import os

# Load an image
path="test"


for img in os.listdir(path):
    # if img == "08e55c5a255b01a4f60c3993241d7f06-full.jpg":
    image_path = os.path.join(path, img)  # Replace with actual image path
    image = Image.open(image_path).convert("RGB")  # Convert to RGB to ensure consistency

    # Apply transformations
    image = transform(image).unsqueeze(0)  # Add batch dimension


    # Perform inference
    with torch.no_grad():
        output = model(image)
        prediction = torch.argmax(output, dim=1).item()

    # Print the result
    classes = ["normal", "porn"]
    print(f"image {img} classified as {classes[prediction]}")
```
